# 주피터 노트북 사용법
* shift + Enter 키를 누르면 셀이 실행되고 커서가 다음셀로 이동합니다.
* Enter 키를 누르면 다시 편집상태로 돌아옵니다.
* ESC 키를 누르고
    * a 키를 누르면 위에 셀이 추가됩니다.
    * b 키를 누르면 아래에 셀이 추가됩니다.
    * dd 키를 누르면 셀이 삭제 됩니다.
    * m 키를 누르면 문서 셀로 변경됩니다.
    * y 키를 누르면 코드 셀로 변경됩니다.
    * h 키를 눌러 단축버튼을 확인할 수 있습니다.

## 마크다운(Markdown)이란?
* 코드와 함께 문서화를 할 수 있습니다.
* 문서화를 할 수 있는 `문법`입니다.

```
여러 줄의 설명을  
줄바꿈으로 쓰고자 할 때  
```
